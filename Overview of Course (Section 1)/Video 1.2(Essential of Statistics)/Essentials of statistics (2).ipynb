Now, let's put some outliers (extreme values) into **R&D Spend** and see the difference.

```python
df['R&D Spend'].max()
```

```python
# Here we define a method to detect the outlier indices
# We will be covering it with much details in upcoming sections
def outlier(df):
    twenty_five= np.percentile(df, 25)
    seventy_five= np.percentile(df, 75)
    iqr= seventy_five - twenty_five
    maximum= seventy_five + 1.5*iqr
    minimum= twenty_five - 1.5*iqr
    return df[df > maximum].index, df[df < minimum].index
```

```python
max_ind, min_ind= outlier(df['R&D Spend'])  # getting outlier indices for R&D Spend
```

```python
print(len(max_ind))
print(len(min_ind))
```

### <font color= 'blue'>Here we can see the number of outliers on both side (both max extreme values and min extreme values are zero). Therefore we saw a very small difference between the mean and median.</font>

Now let's look at the distribution plot.

```python
sns.distplot(df[['R&D Spend']])
```

Here we can see the distribution also looks **normal**. We have explored about normal distribution below.

### <font color= 'green'>Due to lack of outliers and having a nearly normal distribution, the mean and mode of the features are nearly same. And if the distribution would have been NORMAL completely, then mean = median.</font>


-------------------
 Now, let's add some extreme values and see the difference.

```python
# we make a copy of the column and add some higher values
outlier_col= pd.Series(np.append(df['R&D Spend'], [250000, 240000, 280000, 300000, 290000]))
```

```python
# let's check for outliers
max_ind, min_ind = outlier(outlier_col)
print(max_ind)
```

```python
outlier_col.mean()
```

```python
outlier_col.median()
```

```python
sns.distplot(outlier_col)
plt.axvline(x=outlier_col.mean(), color='r', label='mean')
plt.axvline(x=outlier_col.median(), color='g', label='median')
plt.legend()
```

So, we can see that after adding outliers, the graph does not look normal anymore. As well as we can see the difference between mean and median.

###  <font color= 'blue'>While there is very less change in MEDIAN from previous values, the MEAN has undergone a huge change of around 18000. So, when the distribution is not normal, MEDIAN is a better representation of central tendency.

# Inferential Statistics

## <font color= 'blue'>Consider the same example of finding the average weight of people in your society, in Inferential Statistics, you will take a sample set of the people,which is basically a few people from the entire society randomly. You already have grouped the weigth class in maximum,minimum & average. In this method, you basically build a statistical model and expand it for the entire population based on few samples.</font>

### Population
#### <font color= 'green'>A population includes all members from a specified group, all possible outcomes or measurements that are of interest.</font>

### Sample
#### <font color= 'blue'>It is a subset of the population generally formed by random selection of values</font>



# <font color= 'green'>Did you understand the Analogy behind Sampling & why it is needed?</font>
#### Need of sampling
- *The population is too large*.<br> So, applying statistics on a very large data will take more time and computation cost.
<br><br>
- *The population is unreachable*. <br>For example, if we want to get the salaries of engineers of India, then it will be really difficult to reach every individual and collect data. Rather than this, we can collect data from random group of engineers from nearby firms.
<br><br>
- With an appropriate methodology and a sufficiently large sample size,  the results obtained on a sample are often almost as accurate as those that would be obtained on the entire population.
